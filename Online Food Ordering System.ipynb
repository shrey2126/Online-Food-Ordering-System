```python
## import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
import random
import string
import time
import re
import os

# Menu
menu = {
    'Appetizers': {
        'Vegetable Spring Rolls': {'price': 499.00, 'description': 'Crispy spring rolls filled with fresh vegetables.', 'tags': ['Vegan']},
        'Paneer Tikka': {'price': 649.00, 'description': 'Grilled cottage cheese marinated in spices.', 'tags': ['Vegetarian']},
        'Hummus with Pita Bread': {'price': 399.00, 'description': 'Creamy hummus served with warm pita bread.', 'tags': ['Vegan', 'Gluten-Free']},
        'Stuffed Mushrooms': {'price': 549.00, 'description': 'Mushrooms stuffed with cheese and herbs.', 'tags': ['Vegetarian']},
    },
    'Main Courses': {
        'Vegetable Stir-Fry': {'price': 899.00, 'description': 'Fresh vegetables stir-fried in a savory sauce.', 'tags': ['Vegan']},
        'Paneer Butter Masala': {'price': 1049.00, 'description': 'Paneer cooked in a rich tomato and butter sauce.', 'tags': ['Vegetarian']},
        'Mushroom Risotto': {'price': 1199.00, 'description': 'Creamy risotto with mushrooms and Parmesan.', 'tags': ['Vegetarian']},
        'Chickpea Curry': {'price': 999.00, 'description': 'Spicy chickpea curry served with rice.', 'tags': ['Vegan', 'Gluten-Free']},
        'Vegetable Biryani': {'price': 1099.00, 'description': 'Fragrant rice cooked with vegetables and spices.', 'tags': ['Vegan']},
        'Eggplant Parmesan': {'price': 1169.00, 'description': 'Baked eggplant with tomato sauce and cheese.', 'tags': ['Vegetarian']},
    },
    'Salads': {
        'Greek Salad': {'price': 749.00, 'description': 'Fresh vegetables with feta cheese and olives.', 'tags': ['Vegetarian']},
        'Quinoa Salad': {'price': 849.00, 'description': 'Quinoa with vegetables and a lemon vinaigrette.', 'tags': ['Vegan', 'Gluten-Free']},
        'Caesar Salad (Vegetarian)': {'price': 649.00, 'description': 'Romaine lettuce with croutons and Parmesan.', 'tags': ['Vegetarian']},
    },
    'Sides': {
        'Garlic Naan': {'price': 249.00, 'description': 'Soft naan bread with garlic butter.', 'tags': ['Vegetarian']},
        'Steamed Jasmine Rice': {'price': 219.00, 'description': 'Fragrant jasmine rice steamed to perfection.', 'tags': ['Vegan', 'Gluten-Free']},
        'Vegetable Samosas': {'price': 399.00, 'description': 'Crispy samosas filled with spiced vegetables.', 'tags': ['Vegetarian']},
        'Sweet Potato Fries': {'price': 329.00, 'description': 'Crispy sweet potato fries with dipping sauce.', 'tags': ['Vegan']},
    },
    'Beverages': {
        'Fresh Mango Lassi': {'price': 349.00, 'description': 'Refreshing yogurt-based mango drink.', 'tags': ['Vegetarian']},
        'Iced Lemonade': {'price': 249.00, 'description': 'Chilled lemonade with a hint of mint.', 'tags': ['Vegan']},
        'Masala Chai': {'price': 219.00, 'description': 'Spiced Indian tea with milk.', 'tags': ['Vegetarian']},
    },
    'Desserts': {
        'Vegan Chocolate Cake': {'price': 469.00, 'description': 'Rich chocolate cake made without dairy.', 'tags': ['Vegan']},
        'Gulab Jamun': {'price': 429.00, 'description': 'Soft milk-based dumplings in sugar syrup.', 'tags': ['Vegetarian']},
        'Mango Sorbet': {'price': 319.00, 'description': 'Refreshing mango sorbet.', 'tags': ['Vegan']},
    },
    'Combo Deals': {
        'Vegetarian Combo Meal': {'price': 1349.00, 'description': 'A complete meal with appetizer, main course, and dessert.', 'tags': ['Vegetarian']},
        'Paneer Combo Meal': {'price': 1499.00, 'description': 'Paneer-based meal with naan and dessert.', 'tags': ['Vegetarian']},
        'Biryani Meal': {'price': 1249.00, 'description': 'Vegetable biryani with raita and dessert.', 'tags': ['Vegan']},
    },
    'Specials': {
        'Vegan Burger with Fries': {'price': 899.00, 'description': 'Plant-based burger with crispy fries.', 'tags': ['Vegan']},
        'Stuffed Bell Peppers': {'price': 1049.00, 'description': 'Bell peppers stuffed with rice and vegetables.', 'tags': ['Vegan']},
    }
}

# Simulate a user database
users_db = {}

# Cart to store ordered items
cart = {}

# User data storage
current_user = None

# Order History storage
order_history = {}

# File to store users
USERS_FILE = "users.txt"

#file to store order_history
ORDER_HISTORY_FILE = "order_history.txt"

def save_order_history():
    with open(ORDER_HISTORY_FILE, "a") as file:  # Use "a" mode to append to the file
        for username, orders in order_history.items():
            for order in orders:
                # Convert items dictionary to a string format (item1:quantity1,item2:quantity2)
                items_str = ",".join([f"{item}:{qty}" for item, qty in order['items'].items()])
                # Write the order to the file
                file.write(f"{username}|{order['order_id']}|{items_str}|{order['total']}|{order['order_date']}\n")

def load_order_history():
    if os.path.exists(ORDER_HISTORY_FILE):
        with open(ORDER_HISTORY_FILE, "r") as file:
            for line in file:
                line = line.strip()
                if line:
                    username, order_id, items_str, total, order_date = line.split("|")
                    # Convert items string back to a dictionary
                    items = {}
                    for item_qty in items_str.split(","):
                        item, qty = item_qty.split(":")
                        items[item] = int(qty)
                    # Add the order to the order_history dictionary
                    if username not in order_history:
                        order_history[username] = []
                    order_history[username].append({
                        'order_id': order_id,
                        'items': items,
                        'total': float(total),
                        'order_date': order_date
                    })

# Load users from file at startup
def load_users():
    if os.path.exists(USERS_FILE):
        with open(USERS_FILE, "r") as file:
            for line in file:
                username, password, user_id, loyalty_points = line.strip().split("|")
                users_db[username] = {
                    'password': password,
                    'user_id': user_id,
                    'loyalty_points': int(loyalty_points),
                    'orders': []
                }

# Save users to file
def save_users():
    with open(USERS_FILE, "w") as file:
        for username, data in users_db.items():
            file.write(f"{username}|{data['password']}|{data['user_id']}|{data['loyalty_points']}\n")

# Function to display the main menu
def display_menu():
    print("\n--- Welcome to the Vegetarian Food Ordering System ---")
    print("1. View Menu")
    print("2. Search Menu")
    print("3. Take Order")
    print("4. View Cart")
    print("5. Edit Cart")
    print("6. Checkout")
    print("7. View Order History")
    print("8. Visualize Cart")
    print("9. Visualize Order History")
    print("10. Logout")
    print("11. Exit")

# Function to validate the password
def validate_password(password):
    if len(password) < 8:
        return False, "Password must be at least 8 characters long."
    if not re.search(r'[A-Z]', password):
        return False, "Password must contain at least one uppercase letter."
    if not re.search(r'[a-z]', password):
        return False, "Password must contain at least one lowercase letter."
    if not re.search(r'[0-9]', password):
        return False, "Password must contain at least one digit."
    if not re.search(r'[@$!%*?&]', password):
        return False, "Password must contain at least one special character (e.g., !, @, #, $, etc.)."
    return True, "Password is valid."

# Function to register a new user
def register():
    username = input("Enter a new username: ")
    if username in users_db:
        print("Username already exists. Please choose a different username.")
        return
    while True:
        password = input("Enter a new password: ")
        is_valid, message = validate_password(password)
        if is_valid:
            break
        else:
            print(message)
    user_id = generate_order_id()
    users_db[username] = {'password': password, 'user_id': user_id, 'loyalty_points': 0, 'orders': []}
    save_users()  # Save the new user to the file
    print(f"Registration successful. Welcome {username}!")

# Function to log in an existing user
def login():
    global current_user
    username = input("Enter username: ")
    password = input("Enter password: ")
    if username in users_db and users_db[username]['password'] == password:
        current_user = username
        print(f"Welcome back, {username}!")
    else:
        print("Invalid credentials. Please try again.")

# Function to display the entire menu
def view_menu():
    print("\n--- Vegetarian Food Menu ---")
    for category, items in menu.items():
        print(f"\n{category}:")
        for item, details in items.items():
            print(f"  {item}: ₹{details['price']:.2f}")
            print(f"     Description: {details['description']}")
            print(f"     Tags: {', '.join(details['tags'])}")

# Function to search the menu
def search_menu():
    query = input("Enter the name of the item you want to search for: ").lower()
    results = []
    for category, items in menu.items():
        for item, details in items.items():
            if query in item.lower():
                results.append((category, item, details))
    if results:
        print("\nSearch Results:")
        for category, item, details in results:
            print(f"  {item} ({category}): ₹{details['price']:.2f}")
            print(f"     Description: {details['description']}")
            print(f"     Tags: {', '.join(details['tags'])}")
    else:
        print("No items found matching your search.")

# Function to visualize the cart
def visualize_cart():
    if not current_user:
        print("Please log in first to view your cart.")
        return
    if cart:
        item_names = list(cart.keys())
        item_quantities = list(cart.values())
        colors = plt.cm.Paired(np.linspace(0, 1, len(item_names)))
        plt.figure(figsize=(6, 6))
        plt.pie(item_quantities, labels=item_names, autopct='%1.1f%%', startangle=140, colors=colors)
        plt.title(f"Cart Breakdown for {current_user}", fontsize=16, fontweight='bold')
        plt.axis('equal')
        plt.show()
    else:
        print("\nYour cart is empty.")

# Function to visualize order history
def visualize_order_history():
    if not current_user:
        print("Please log in first to view your order history.")
        return
    if current_user in order_history and order_history[current_user]:
        # Group orders by date
        date_spending = {}
        for order in order_history[current_user]:
            # Extract the date from the order_id
            try:
                # Split the order_id by underscores and extract the date part
                order_date = order['order_id'].split('_')[1]  # Example: "ABC12345_2023-10-15"
            except IndexError:
                # If the order_id does not contain a date, skip this order
                print(f"Skipping order {order['order_id']}: Invalid format (missing date).")
                continue

            # Add the order total to the corresponding date
            if order_date in date_spending:
                date_spending[order_date] += order['total']
            else:
                date_spending[order_date] = order['total']

        # Check if there are any valid orders with dates
        if not date_spending:
            print("No valid orders with dates found.")
            return

        # Plot total spending per day
        dates = list(date_spending.keys())
        spendings = list(date_spending.values())
        plt.figure(figsize=(12, 5))
        plt.bar(dates, spendings, color='skyblue')
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Total Spending (₹)', fontsize=12)
        plt.title(f"Daily Spending for {current_user}", fontsize=16, fontweight='bold')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    else:
        print("No orders found.")

# Function to take an order
def take_order():
    if not current_user:
        print("Please log in first to place an order.")
        return

    while True:
        print("\nSelect a category to order from:")
        print("1. Appetizers")
        print("2. Main Courses")
        print("3. Salads")
        print("4. Sides")
        print("5. Beverages")
        print("6. Desserts")
        print("7. Combo Deals")
        print("8. Specials")
        print("9. Back to Main Menu")

        category_choice = input("Enter category number (1-9): ")

        if category_choice == '9':
            break
        elif category_choice in ['1', '2', '3', '4', '5', '6', '7', '8']:
            category_map = {
                '1': 'Appetizers', '2': 'Main Courses', '3': 'Salads', '4': 'Sides',
                '5': 'Beverages', '6': 'Desserts', '7': 'Combo Deals', '8': 'Specials'
            }
            category = category_map[category_choice]
            print(f"\nAvailable items in {category}:")
            for item, details in menu[category].items():
                print(f"  {item}: ₹{details['price']:.2f}")
                print(f"     Description: {details['description']}")
                print(f"     Tags: {', '.join(details['tags'])}")

            item_choice = input(f"Enter the item you want to order from {category}: ")

            item_choice_lower = item_choice.lower()
            item_found = False

            for item in menu[category]:
                if item.lower() == item_choice_lower:
                    item_found = True
                    break

            if item_found:
                quantity = int(input(f"Enter quantity for {item_choice}: "))
                if item_choice in cart:
                    cart[item_choice] += quantity
                else:
                    cart[item_choice] = quantity
                print(f"{item_choice} x{quantity} has been added to your cart.")
            else:
                print(f"Sorry, {item_choice} is not available in the {category} category.")
        else:
            print("Invalid choice. Please select a valid category.")

# Function to display the cart contents
def display_cart():
    if not current_user:
        print("Please log in first to view your cart.")
        return

    if cart:
        print("\n--- Your Cart ---")
        total_price = 0
        for item, quantity in cart.items():
            item_price = None
            # Loop through the menu categories to find the item
            for category, items in menu.items():
                # Compare lowercase version of item to ensure case-insensitive matching
                for menu_item in items:
                    if menu_item.lower() == item.lower():  # case-insensitive comparison
                        item_price = items[menu_item]['price']
                        break
                if item_price:
                    break

            if item_price is None:
                print(f"Error: Price for {item} not found in the menu.")
                continue  # Skip this item if no price is found

            item_total = item_price * quantity
            print(f"{item}: {quantity} x ₹{item_price:.2f} = ₹{item_total:.2f}")
            total_price += item_total

        print(f"\nTotal Price: ₹{total_price:.2f}")
    else:
        print("\nYour cart is empty.")

# Function to edit the cart
def edit_cart():
    if not current_user:
        print("Please log in first to edit your cart.")
        return

    if not cart:
        print("Your cart is empty. Add items to the cart before editing.")
        return

    print("\n--- Edit Cart ---")
    for item, quantity in cart.items():
        print(f"{item}: {quantity}")

    item_to_edit = input("Enter the name of the item you want to edit: ").strip()  # Trim spaces
    item_to_edit_lower = item_to_edit.lower()  # Convert to lowercase for case-insensitive comparison

    # Search for the item in the cart (case-insensitive)
    found_item = None
    for item in cart:
        if item.lower() == item_to_edit_lower:  # Case-insensitive comparison
            found_item = item
            break

    if found_item:
        new_quantity = int(input(f"Enter the new quantity for {found_item}: "))
        if new_quantity > 0:
            cart[found_item] = new_quantity
            print(f"{found_item} quantity updated to {new_quantity}.")
        else:
            del cart[found_item]
            print(f"{found_item} removed from the cart.")
    else:
        print(f"{item_to_edit} not found in the cart.")

# Function to process the payment based on selected method
def process_payment(payment_method, total_price):
    if payment_method == '1':  # Credit Card
        card_number = input("Enter your Credit Card Number (16 digits): ")
        if len(card_number) == 16 and card_number.isdigit():
            print(f"Payment of ₹{total_price:.2f} processed successfully using Credit Card.")
            return True
        else:
            print("Invalid Credit Card number. Please try again.")
            return False

    elif payment_method == '2':  # Debit Card
        card_number = input("Enter your Debit Card Number (16 digits): ")
        if len(card_number) == 16 and card_number.isdigit():
            print(f"Payment of ₹{total_price:.2f} processed successfully using Debit Card.")
            return True
        else:
            print("Invalid Debit Card number. Please try again.")
            return False

    elif payment_method == '3':  # Net Banking
        bank_name = input("Enter your Bank Name: ")
        if bank_name:
            print(f"Payment of ₹{total_price:.2f} processed successfully using Net Banking.")
            return True
        else:
            print("Invalid Bank Name. Please try again.")
            return False

    else:
        print("Invalid Payment Method. Please try again.")
        return False

def validate_upi(upi_id):
    # Regular expression to match UPI ID format
    upi_regex = r'^[a-zA-Z0-9_.]+@[a-zA-Z0-9]+\.[a-zA-Z]{2,}$'
    
    if re.match(upi_regex, upi_id):
        return True
    else:
        print("Invalid UPI ID format. Please ensure it follows the format <username>@<bank_domain>.")
        return False

# Function to generate a custom order ID of 8-9 characters
def generate_order_id():
    # Generate a random string of 8 characters
    random_string = ''.join(random.choices(string.ascii_letters + string.digits, k=8))
    # Add the current date to the order ID
    order_date = datetime.now().strftime("%Y-%m-%d")
    return f"{random_string}_{order_date}"

# Function to generate a bill
def generate_bill(order_id, total_price, order_items, discount=0, discount_code=None):
    import datetime

    # Get current date and time
    now = datetime.datetime.now()
    date_time = now.strftime("%Y-%m-%d %H:%M:%S")
    
    # Prepare the bill content
    total_amount = max(0, total_price - discount)  # Ensure the total is not negative
    bill_content = f"--- Bill ---\nOrder ID: {order_id}\nDate: {date_time}\n\nItems:\n"
    
    for item, quantity in order_items.items():
        item_price = None
        # Loop through menu categories to find the item
        for category, items in menu.items():
            for menu_item in items:
                if menu_item.lower() == item.lower():  # Case-insensitive comparison
                    item_price = items[menu_item]['price']  # Get the price of the item
                    break
            if item_price is not None:
                break  # Exit the loop once the item is found
        
        if item_price is not None:
            item_total = item_price * quantity
            bill_content += f"{item}: {quantity} x ₹{item_price:.2f} = ₹{item_total:.2f}\n"
        else:
            bill_content += f"{item}: Price not found in the menu.\n"  # If price is not found

    # Add discount details to the bill
    if discount_code:
        bill_content += f"\nDiscount Applied: {discount_code} (-₹{discount:.2f})"
    bill_content += f"\nTotal: ₹{total_amount:.2f}"  # Display the final total after discount

    # Save the bill to a text file with UTF-8 encoding
    bill_filename = f"bill_{order_id}.txt"
    with open(bill_filename, 'w', encoding='utf-8') as file:
        file.write(bill_content)
    
    print(f"Bill has been generated and saved as {bill_filename}")

# Function to apply discounts
def apply_discount(total_price, discount_code, loyalty_points):
    discount = 0
    if discount_code == "SAVE10":  # 10% discount
        discount = 0.10 * total_price
        print(f"Discount applied: ₹{discount:.2f}")
    elif discount_code == "SAVE20":  # 20% discount
        discount = 0.20 * total_price
        print(f"Discount applied: ₹{discount:.2f}")
    elif discount_code == "FREESHIP":  # Flat ₹100 discount
        discount = 100
        print(f"Discount applied: ₹{discount:.2f}")
    elif discount_code == "LOYAL50" and loyalty_points >= 50:  # 50% discount for loyal customers
        discount = 0.50 * total_price
        print(f"Discount applied: ₹{discount:.2f}")
    elif discount_code == "LOYAL50" and loyalty_points < 50:
        print("You need at least 50 loyalty points to use this coupon.")
    else:
        print("Invalid or expired discount code.")
    return discount

# Modified checkout function to include bill generation and loyalty points
def checkout():
    if not current_user:
        print("Please log in first to checkout.")
        return

    if not cart:
        print("Your cart is empty. Add items to the cart before checkout.")
        return

    print("\n--- Checkout ---")
    
    # Calculate total price from the cart
    total_price = 0
    for item, quantity in cart.items():
        item_found = False
        # Loop through menu categories to find the item
        for category, items in menu.items():
            for menu_item in items:
                # Compare lowercase version of item to ensure case-insensitive matching
                if menu_item.lower() == item.lower():  # case-insensitive comparison
                    item_price = items[menu_item]['price']  # Get the price of the item
                    item_total = item_price * quantity
                    print(f"{item}: {quantity} x ₹{item_price:.2f} = ₹{item_total:.2f}")
                    total_price += item_total
                    item_found = True
                    break
            if item_found:
                break
        
        if not item_found:
            print(f"Error: {item} not found in the menu.")

    print(f"\nTotal Price: ₹{total_price:.2f}")
    
    # Apply discount if applicable
    discount_code = input("Enter a discount code (if any, otherwise press Enter): ")
    discount = 0
    if discount_code:
        discount = apply_discount(total_price, discount_code, users_db[current_user]['loyalty_points'])
    total_price -= discount
    print(f"New Total Price: ₹{total_price:.2f}")

    # Proceed with payment
    print("\nChoose Payment Method:")
    print("1. Credit Card")
    print("2. Debit Card")
    print("3. UPI")
    print("4. Net Banking")
    print("5. Cash on Delivery")

    payment_method = input("Enter payment method (1-5): ")

    if payment_method in ['1', '2', '3', '4', '5']:
        if payment_method == '1':  # Credit Card
            card_number = input("Enter your Credit Card Number (16 digits): ")
            if len(card_number) == 16 and card_number.isdigit():
                print(f"Payment of ₹{total_price:.2f} processed successfully using Credit Card.")
                return True
            else:
                print("Invalid Credit Card number. Please try again.")
                return False

        if payment_method == '2':  # Debit Card
            card_number = input("Enter your Debit Card Number (16 digits): ")
            if len(card_number) == 16 and card_number.isdigit():
                print(f"Payment of ₹{total_price:.2f} processed successfully using Debit Card.")
                return True
            else:
                print("Invalid Debit Card number. Please try again.")
                return False
        if payment_method == '3':  # UPI Payment
            upi_id = input("Enter your UPI ID (e.g., example@upi): ")
            if not validate_upi(upi_id):
                print("Invalid UPI ID format. Please try again.")
                return
        if payment_method == '4':  # Net Banking
            bank_name = input("Enter your Bank Name: ")
            account_number = input("Enter your Bank Account Number: ")
            if not account_number.isdigit():
                print("Invalid account number. Please try again.")
                return

        # Generate order ID and date
        order_id = generate_order_id()
        order_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        # Add the order to the order history
        if current_user not in order_history:
            order_history[current_user] = []
        order_history[current_user].append({
            'order_id': order_id,
            'items': cart.copy(),
            'total': total_price,
            'order_date': order_date
        })

        # Save the order history to the file
        save_order_history()

        # Generate the bill
        generate_bill(order_id, total_price, cart, discount, discount_code)

        # Add loyalty points
        loyalty_points = int(total_price // 100)
        users_db[current_user]['loyalty_points'] += loyalty_points
        save_users()  # Save the updated loyalty points to users.txt

        print(f"Order placed successfully. Your Order ID is {order_id}. Thank you for your purchase!")
        print(f"You earned {loyalty_points} loyalty points!")
        cart.clear()  # Clear the cart after checkout
    else:
        print("Invalid payment method selected. Please try again.")

# Function to display order history
def view_order_history():
    if not current_user:
        print("Please log in first to view order history.")
        return

    if current_user in order_history and order_history[current_user]:
        print(f"\n--- Order History for {current_user} ---")
        for order in order_history[current_user]:
            print(f"\nOrder ID: {order['order_id']}")
            print(f"Order Date: {order['order_date']}")
            print("Items:")
            for item, quantity in order['items'].items():
                item_price = None
                # Loop through menu categories to find the item, case-insensitive
                for category, items in menu.items():
                    for menu_item in items:
                        if menu_item.lower() == item.lower():  # Case-insensitive comparison
                            item_price = items[menu_item]['price']  # Get the price of the item
                            break
                    if item_price is not None:
                        break  # Exit the loop once the item is found
                
                if item_price is not None:
                    item_total = item_price * quantity
                    print(f"  {item}: {quantity} x ₹{item_price:.2f} = ₹{item_total:.2f}")
                else:
                    print(f"  {item}: Price not found in the menu.")  # If price is not found

            print(f"Total: ₹{order['total']:.2f}")
    else:
        print("No orders found.")

# Main function to control the flow of the program
def main():
    global current_user

    while True:
        display_menu()

        choice = input("Enter your choice (1-11): ")

        if choice == '1':
            view_menu()
        elif choice == '2':
            search_menu()
        elif choice == '3':
            if current_user:
                take_order()
            else:
                print("Please log in to place an order.")
        elif choice == '4':
            display_cart()
        elif choice == '5':
            edit_cart()
        elif choice == '6':
            checkout()
        elif choice == '7':
            view_order_history()
        elif choice == '8':
            visualize_cart()
        elif choice == '9':
            visualize_order_history()
        elif choice == '10':
            if current_user:
                print(f"{current_user} logged out successfully.")
                current_user = None
            else:
                print("You are not logged in.")
        elif choice == '11':
            print("Thank you for visiting! Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the main function
if __name__ == "__main__":
    load_users()  # Load users from file at startup
    load_order_history()  # Load order history from file at startup
    while True:
        print("\n--- Welcome to the Vegetarian Food Ordering System ---")
        print("1. Register")
        print("2. Login")
        print("3. Exit")

        option = input("Choose an option: ")

        if option == '1':
            register()
        elif option == '2':
            login()
            main()  # Start the main menu after login
        elif option == '3':
            print("Thank you for visiting! Goodbye!")
            break
        else:
            print("Invalid option. Please choose again.")
```
